### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
df['LanguageHaveWorkedWith'].unique()
```

```python
df['LanguageWantToWorkWith'].unique()
```

```python
df['LanguageHaveWorkedWith'].describe()
```

```python
df['LanguageHaveWorkedWith'].info()
```

```python
df['LanguageWantToWorkWith'].info()
```

```python
df.dropna(subset=['LanguageHaveWorkedWith','LanguageWantToWorkWith'],inplace=True)
```

```python
df['LanguageHaveWorkedWith']=df['LanguageHaveWorkedWith'].str.split(';')
```

```python
df['LanguageWantToWorkWith']=df['LanguageWantToWorkWith'].str.split(';')
```

```python
codes_worked_dic = {}
for codes_work in df['LanguageHaveWorkedWith']:
    for code in codes_work:
        code =code.strip()
        codes_worked_dic[code] = codes_worked_dic.get(code,0)+1
```

```python
codes_worked_dic = dict(sorted(codes_worked_dic.items(), key=lambda item:item[1],reverse=True))            
```

```python
codes_want_dic = {}
for codes_want in df['LanguageWantToWorkWith']:
    for code in codes_want:
        code =code.strip()
        codes_want_dic[code] = codes_want_dic.get(code,0)+1
```

```python
codes_want_dic=dict(sorted(codes_want_dic.items(),key=lambda item:item[1],reverse=True))
```

```python
import copy
codes_compare=copy.deepcopy(codes_worked_dic)
for code in codes_want_dic.keys():
    codes_compare[code]=[codes_compare.get(code,0),codes_want_dic.get(code)]
```

```python
codes_compare
```

```python
ax=plt.figure(figsize=(14,10))
plt.barh(codes_worked_dic.keys(),codes_worked_dic.values(),color='blue',label='Worked With',alpha=0.4)
plt.barh(codes_want_dic.keys(),codes_want_dic.values(),color='red',label= 'Want work with',alpha=0.6)
plt.legend()
plt.xlabel('Number of Person')
plt.ylabel('Coding Language')
plt.title('Comparsion Between Current Work and Want Work with Coding Language')
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
df['RemoteWork'].info()
```

```python
df['RemoteWork'].describe()
```

```python
df['Country'].info()
```

```python
df['Country'].describe()
```

```python
pivot_table = pd.crosstab(df['Country'],df['RemoteWork'])
```

```python
top10_countries= pivot_table.sum(axis=1).sort_values(ascending=False).head(10).index
```

```python
pivot_top10 = pivot_table.loc[top10_countries]
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(pivot_top10,cmap='viridis',annot=True,annot_kws={'size':12},fmt='d', linewidths=0.5)
plt.title('Distribution of Remote Work by Country', fontsize=16)  
plt.ylabel('Country', fontsize=12)  
plt.xlabel('Remote Work Type', fontsize=12) 
plt.tight_layout()
plt.show()
```

```python
# 步骤1：导入必要的库  
import pandas as pd  
import matplotlib.pyplot as plt  
import seaborn as sns  

# 步骤2：对国家进行分组（可选，如果国家太多）  
# 如果不需要分组，可以跳过这一步  
# 例如：创建一个字典将国家映射到区域  
# region_mapping = {...}  
# df['Region'] = df['Country'].map(region_mapping)  

# 步骤3：创建数据透视表  
pivot_table = pd.crosstab(df['Country'], df['RemoteWork'])  

# 步骤4：选择排名前N的国家（可选，如果国家太多）  
top_countries = pivot_table.sum(axis=1).sort_values(ascending=False).head(20).index  
pivot_filtered = pivot_table.loc[top_countries]  

# 步骤5：绘制热力图  
plt.figure(figsize=(12, 10))  
sns.heatmap(pivot_filtered, cmap='viridis', annot=True, fmt='d', linewidths=.5)  

# 步骤6：设置图表标题和标签  
plt.title('Distribution of Remote Work by Country', fontsize=16)  
plt.ylabel('Country', fontsize=12)  
plt.xlabel('Remote Work Type', fontsize=12)  

# 步骤7：显示图表  
plt.tight_layout()  
plt.show()  
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
df[['JobSat','YearsCodePro']].info()
```

```python
df[['JobSat','YearsCodePro']].describe(include='all')
```

```python
df['YearsCodePro'].unique()
df['YearsCodePro'].replace({'Less than 1 year':0.5,'More than 50 years':60},inplace=True)
```

```python
df['YearsCodePro']=df['YearsCodePro'].astype(float)
```

```python
df['JobSat']=df['JobSat'].astype(int)
df['JobSat'].unique()
```

```python
df.dropna(subset=['JobSat','YearsCodePro'],inplace=True)
```

```python
from scipy import stats
pearson_coef,p_value =stats.pearsonr(df['JobSat'],df['YearsCodePro'])
print(f'pearson correlation: {pearson_coef:.4f},p value:{p_value:.4f}')
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
df[['Employment','EdLevel']].info()
```

```python
df[['Employment','EdLevel']].describe()
```

```python
df['EdLevel'].unique()
```

```python
df_Ed_Emp=df[['Employment','EdLevel']].explode('Employment')
```

```python
df_Ed_Emp.info()
```

```python
df_Ed_Emp.describe()
```

```python
pivot_Emp_Ed = pd.crosstab(df_Ed_Emp['Employment'],df_Ed_Emp['EdLevel'])
plt.figure(figsize=(16, 10))  
pivot_Emp_Ed.plot(kind='bar', stacked=True)  
plt.xlabel('Employment Status')  
plt.ylabel('Number of People with different Education Level')  
plt.title('Distribution of different education level across Employment Status')  
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)  # 图例放在右侧  
plt.subplots_adjust(right=0.75)  # 调整右边距  
plt.show()  
```
